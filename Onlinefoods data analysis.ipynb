```python
plt.figure(figsize=(8,6))
plt.scatter(df['Age'],df['Occupation'],color='purple',alpha=0.5)
plt.title('scatter of Age')
plt.xlabel('Age')
plt.ylabel('Occupation')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5, 5))
plt.hist(df['Gender'],bins=15,color='green',edgecolor='violet')
plt.title('hist Chart Example:Age by gender')
plt.xlabel('Age')
plt.ylabel('Gender')
plt.grid(True)
plt.show()
```
